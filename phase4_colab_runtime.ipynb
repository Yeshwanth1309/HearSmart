```python
# ===== CELL 5: VERIFY PROJECT STRUCTURE =====

import os
import numpy as np

required_paths = [
    "src",
    "features/mfcc/train",
    "features/mfcc/val",
    "features/mfcc/test",
    "data/splits/train.csv",
    "models",
    "results"
]

for path in required_paths:
    if not os.path.exists(path):
        raise FileNotFoundError(f"Missing required path: {path}")
    else:
        print(f"[OK] {path}")

print("\nChecking MFCC counts...")

print("Train:", len(os.listdir("features/mfcc/train")))
print("Val:", len(os.listdir("features/mfcc/val")))
print("Test:", len(os.listdir("features/mfcc/test")))

sample_file = os.listdir("features/mfcc/train")[0]
sample = np.load(f"features/mfcc/train/{sample_file}")

assert sample.shape == (80,), "MFCC shape mismatch!"
print("\nMFCC sample shape verified:", sample.shape)

print("\n===== REMOTE PROJECT VALIDATION PASSED =====")
```
